```python
# Write a program that uses try, except, and finally blocks together.
try:
    a = int(input("Enter the First number: "))
    b = int(input("Enter the Second number: "))
    print(a/b)
except ZeroDivisionError as z:
    print("Error:",z)
except ValueError as v:
    print("Error:",v)
finally:
    print("run always")
```

```python
# Write a program that catches multiple exceptions using multiple except blocks.
try:
    a = int(input("Enter the First number: "))
    b = int(input("Enter the Second number: "))
    print(a/b)
except ZeroDivisionError:
    print("Error Zero division")
except ValueError:
    print("Error Value Error")
except Exception as e:
    print("Unexpected error",e)
```

### PYTHON COMPREHENSIONS
    * Comprehensions = Python ka “shortcut” way to create lists, sets, dictionaries, generators.
    * Same kaam kam code me, fast, readable, optimized.

```python
# List comprehension
numbers = [i for i in range(1,6)]
print(numbers)
```

```python
# square of a number 
squares = [x*x for x in range(1, 6)]
squares
```

```python
# even numbers from 1 to 10.
evens = [i for i in range(1,11) if i%2==0 ]
print(evens)
```

```python
# Print even number from 1 to 10 but where odd number comes, print "odd" (if-else)
result = [i if i%2==0 else "Odd" for i in range(1,11)]
print(result)
```

```python
# python comprehension with lambda
n = [1,2,3,4]
m = lambda d:[i*i for i in n]
print(m(n))
```

```python
b = [i*i for i in range(1,5)]
print(b)
```

```python
n = [1,2,3,4]
m = list(map(lambda d: d*d,n))
print(m)
```

```python
# DICTIONARY COMPREHENSION
# Dictionary : A dictionary is a collection of data stored as key : value pairs.
# Dictionary comprehension: Dictionary comprehension is a short and smart way to create dictionaries using a loop.
```

```python
# Normal dictionary
d = {}
for i in range(1, 6):
    d[i] = i*i
print(d)
```

```python
# with it 
d = {i:i*i for i in range(1,6)}
d
```

```python
# nos. and cubes
cubes = {x: x**3 for x in range(1, 6)}
cubes
```

```python
# SET COMPREHENSION
# Set : A set is a collection of unique values.
# Set comprehension : Set comprehension is a compact way to create a set using a loop.
```

```python
s = {1,12,22,1,1,4,5,3,32,3}
s
```

```python
b = [1,1,1,2,2,2,3,3,3]
s = {x for x in b}
list(s)
```

```python
# remove duplicates
nums = [1, 2, 2, 3, 3, 4]
unique = list({x for x in nums})
unique
```

# Module 
    In Python, modules allow you to organize and reuse your code. A module is simply a .py file that contains functions, classes, or variables that you can use in other programs. Importing modules is a fundamental practice for writing clean, maintainable, and scalable code.
        Python file (.py) which contains functions, classes, variables.
        Example: math.py, os.py
    
### Why Import Modules?
    Reuse: Use already written functions and classes, avoiding repeated code.
    Organization: Divide your project into smaller, more manageable files.

### Importing Standard Modules
    Python includes a large number of ready-to-use modules. For example, to work with dates you can import the datetime module:
    

```python
import datetime as dt
# dir(dt)
today = dt.datetime.now()
print("Today's date  is :" , today)
```

```python
from datetime import datetime
dir(datetime)
```
