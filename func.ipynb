# Recursion
    A function that calls itself to solve a problem by breaking it into smaller parts.
     Used when a big problem can be broken into smaller similar problems.

    Every recursive function MUST have:

        Base case (stop condition)
        
        Recursive call (function calls itself)
### Two IMPORTANT Parts of Recursion
    Every recursive function must have:
    
    1️ Base Case
        Stops the recursion
    
    2️ Recursive Call
        Function calling itself
    
    Without base case → infinite loop

        
### syntax:
    def hello(problem):
        if base_condition:
            return result
        return hello(problem - smaller_problem)
    hello()

```python
# Program to find the factorial of a number using recursion
# 4! = 4 * 3 * 2 * 1 = 24
def factorial(n): # n=4
    if n == 1:   # base case (factorial(1) = 1)
        return 1 # stop the condition or recursion
    return n * factorial(n-1) # recursive call
factorial(4)
# (n=4) factorial(4) =  4* factorial(3) = 4 * 6 = 24
# (n=3) factorial(3) =  3* factorial(2) = 3 * 2 = 6
#(n=2) factorial(2) =  2* factorial(1) = 2 * 1 = 2
#(n=1) factorial(1) = 1
```

```python
def fac(n):
    fact = 1
    for i in range(1,n+1):
        fact *= i
    return fact
fac(5)
```

```python
## print 5 to 1 in reverse order using recursion
# 5 4 3 2 1
def show(n):
    if n == 1:        # base case
        return 1 # stop
    print(n)
    show(n - 1)       # recursive call
show(5)
```

```python
# fabonacci using recursion
def fibonacci(n):
    if n == 0: # base case
        return 0
    elif n == 1:
        return 1 # base case
    else:
        return fibonacci(n-1) + fibonacci(n-2) # recursive call
fibonacci(6)
# normal num = 0 1 2 3 4 5 6
# fabinacci  = 0 1 1 2 3 5 8
# fibonacci(6) = fibonacci(5) + fibonacci(4) = 5 + 3 = 8
# fibonacci(5) = fibonacci(4) + fibonacci(3) = 3 + 2 = 5
# fibonacci(4) = fibonacci(3) + fibonacci(2) = 2 + 1 = 3
# fibonacci(3) = fibonacci(2) + fibonacci(1) = 1 + 1 = 2
# fibonacci(2) = fibonacci(1) + fibonacci(0) = 1 + 0 = 1
# fibonacci(1) = 1
# fibonacci(0) = 0
```

```python
# Sum of first 5 numbers using recursion
# 5 + 4 + 3 + 2 + 1 = 15
def sum_n(n):
    if n == 1:          # base case
        return 1 # stop
    return n + sum_n(n-1) # recursive call
sum_n(5)
# sum_n(5) : 5 + sum_n(4) = 5 + 10 = 15
# sum_n(4) : 4 + sum_n(3) = 4 + 6 = 10
# sum_n(3) : 3 + sum_n(2) = 3 + 3 = 6
# sum_n(2) : 2 + sum_n(1) = 2 + 1 = 3
# sum_n(1) = 1
```

```python
# Mini Project - To Do App

def task():
    tasks = []

    print("----- WELCOME TO THE TASK MANAGEMENT APP -----")

    # Initial tasks
    total_task = int(input("Enter how many tasks you want to add: "))
    for i in range(1, total_task + 1):
        task_name = input(f"Enter task {i}: ")
        tasks.append(task_name)

    print(f"\nToday's tasks are:\n{tasks}")

    # Menu loop
    while True:
        print("\n===== MENU =====")
        print("1. Add Task")
        print("2. Update Task")
        print("3. Delete Task")
        print("4. View Tasks")
        print("5. Exit")

        operation = int(input("Enter your choice: "))

        # ADD TASK
        if operation == 1:
            add = input("Enter task you want to add: ")
            tasks.append(add)
            print(f"Task '{add}' has been successfully added.")

        # UPDATE TASK
        elif operation == 2:
            updated_val = input("Enter the task name you want to update: ")
            if updated_val in tasks:
                up = input("Enter new task: ")
                ind = tasks.index(updated_val)
                tasks[ind] = up
                print(f"Task updated to '{up}'")
            else:
                print("Task not found!")

        # DELETE TASK
        elif operation == 3:
            del_val = input("Enter task you want to delete: ")
            if del_val in tasks:
                tasks.remove(del_val)
                print(f"Task '{del_val}' deleted.")
            else:
                print("Task not found!")

        # VIEW TASKS
        elif operation == 4:
            if len(tasks) == 0:
                print("No tasks available.")
            else:
                print("\nYour Tasks:")
                for i, t in enumerate(tasks, start=1):
                    print(f"{i}. {t}")

        # EXIT
        elif operation == 5:
            print("Thank you for using Task Manager 🙏")
            break

        else:
            print("Invalid choice. Try again.")

# Run the app
task()
```
